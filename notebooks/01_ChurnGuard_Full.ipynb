# ChurnGuard – Telco Customer Retention Engine

## Business Problem
Telecommunication companies often struggle with **customer churn**, where users stop using their service. In this dataset, about **26.5% of customers leave each year**, which creates a major revenue loss.  

Since the **average revenue per user (ARPU) is around $65 per month**, every customer who churns represents nearly **$780 in lost annual revenue**. With over **7,000 customers**, this results in **more than $1.2 million in potential yearly revenue at risk**.

---

## Project Objective
The goal of this project is to **identify customers who are likely to churn and help the company retain them before they leave**.  

By combining **customer segmentation (RFM analysis)** with a **Logistic Regression churn prediction model**, the project aims to reduce churn from **26.5% to around 11.5%**, which could protect roughly **$187,000 in annual revenue**.

---

## Approach

### 1. Customer Segmentation (RFM Analysis)
Customers are grouped based on their behavior and engagement into segments such as:

- Champions  
- Loyal Customers  
- At-Risk Customers  
- Hibernating Customers  
- Lost Customers  

This helps the business understand **which customers are valuable and which ones are likely to churn**.

### 2. Churn Prediction Model
A **Logistic Regression model** predicts the probability that each customer will churn.  
This allows the company to **focus retention efforts on customers with the highest risk of leaving**.

---

## Retention Strategy
Based on the customer segment and churn probability, targeted retention strategies can be applied, such as:

- Discounts or promotional offers  
- Service upgrades or loyalty benefits  
- Personalized win-back campaigns for inactive customers  

These actions help **increase customer retention and reduce revenue loss**.

---

## Results
- **Model Accuracy:** 79%  
- **Retention improvement in the At-Risk segment:** 45%  
- **Estimated annual revenue protected:** ~$187,000

## Resume Version
**ChurnGuard – Telco Customer Retention System**  

Developed an end-to-end churn prediction and retention framework using **RFM segmentation and Logistic Regression** on a dataset of **7,043 telecom customers**. Achieved **79% prediction accuracy** and designed segment-based retention strategies projected to **protect ~$187K in annual revenue**.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/Telco-Customer-Churn.csv")

# Basic validation
print("Dataset Shape:", df.shape)

churn_rate = df['Churn'].value_counts(normalize=True)['Yes']
print("Churn Rate:", round(churn_rate * 100, 2), "%")

arpu = df['MonthlyCharges'].mean()
print("Average Monthly ARPU: $", round(arpu, 2))

# Revenue risk calculation
total_customers = len(df)
annual_revenue_at_risk = total_customers * churn_rate * arpu * 12

print(f"Annual Revenue at Risk: ${annual_revenue_at_risk:,.0f}")

# If churn reduced by 15%
projected_savings = annual_revenue_at_risk * 0.15
print(f"Projected Savings after Churn Reduction: ${projected_savings:,.0f}")
```
